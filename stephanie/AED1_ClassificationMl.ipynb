## üìå Analyse des valeurs manquantes

Lors de l‚Äôanalyse des donn√©es, on observe que trois colonnes pr√©sentent environ **93 % de valeurs manquantes** :

- `date_retour`
- `motif`
- `mois_retour`

Ces colonnes sont toutes li√©es √† la **gestion des retours**.

üéØ **Explication :**  
La grande majorit√© des commandes **n‚Äôont pas √©t√© retourn√©es**, ce qui est logique dans un processus de vente classique.  
Ainsi, les champs relatifs au retour (date, motif, mois) restent **vides** pour ces commandes.

‚úÖ **Conclusion :**  
Ces valeurs manquantes ne sont pas des erreurs, mais refl√®tent simplement le fait que **seules une minorit√© de commandes (environ 7 %) ont donn√© lieu √† un retour produit**.  
On pourra donc traiter ces champs avec prudence lors des analyses (ex. : filtrer uniquement les lignes avec retour quand n√©cessaire).


```python

# ‚úÖ Objectif : traiter les colonnes li√©es aux retours

# 1. Cr√©er une copie du DataFrame original
data_ml = data.copy()

# 2. Cr√©er une colonne binaire "retour" :
#    - 1 si la commande a √©t√© retourn√©e (date_retour non nulle)
#    - 0 sinon
data_ml['retour'] = data_ml['date_retour'].notnull().astype(int)

# 3. Remplacer les valeurs manquantes :
#    - 'motif' ‚ûú "Aucun"
#    - 'mois_retour' ‚ûú "Non retourn√©"
data_ml['motif'] = data_ml['motif'].fillna("Aucun")
data_ml['mois_retour'] = data_ml['mois_retour'].fillna("Non retourn√©")

# 4. Supprimer la colonne "date_retour" (non num√©rique, peu utile pour mod√©lisation)
data_ml.drop(columns=['date_retour'], inplace=True)

# 5. V√©rification du r√©sultat sur les premi√®res lignes
data_ml[['retour', 'motif', 'mois_retour']].head()
```

Apr√®s suppression des doublons exacts, nous avons observ√© que **44 identifiants de commande (`id_commande`)** apparaissaient encore plusieurs fois.

üîé **Observation d√©taill√©e :**
- En examinant les lignes associ√©es √† ces `id_commande`, nous avons constat√© qu'elles √©taient **totalement identiques** (m√™me produit, m√™me quantit√©, m√™me entrep√¥t, etc.).

üéØ **Action :**

```python
# üìå √âtape : Suppression des doublons sur id_commande

# ‚úÖ Objectif : Ne conserver qu'une seule ligne par "id_commande" pour √©viter les r√©p√©titions injustifi√©es

# Suppression en gardant la premi√®re occurrence de chaque id_commande
data_ml = data_ml.drop_duplicates(subset=['id_commande'], keep='first')

# V√©rification : il ne doit plus y avoir de doublons sur id_commande
doublons_id_commande_apres = data_ml['id_commande'].duplicated().sum()
print(f"Nombre de doublons sur id_commande apr√®s suppression : {doublons_id_commande_apres}")
```

On s√©lectionne les colonnes utiles pour la pr√©diction, puis on encode les variables cat√©gorielles (cat√©gorie, marque, entrep√¥t) avec `get_dummies`.
Ensuite, on divise les donn√©es en un ensemble d'entra√Ænement et un ensemble de test (80/20).


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# üéØ Cible
y = data_ml["retour"]

# üßÆ Variables explicatives (√† ajuster si besoin)
X = data_ml[[
    "prix", "quantit√©_command√©e", "stock_minimum", "niveau_stock",
    "cat√©gorie", "marque", "entrepot_commande"
]]

# üè∑Ô∏è Encodage des variables cat√©gorielles (one-hot)
X_encoded = pd.get_dummies(X)

# üìÇ Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)
```

## Algorithme de ml Regression logistic

### ‚öñÔ∏è √âtape 1 : Normalisation
La r√©gression logistique n√©cessite une mise √† l‚Äô√©chelle des variables num√©riques.
On applique `StandardScaler` uniquement aux colonnes num√©riques.


```python
from sklearn.preprocessing import StandardScaler

# On normalise seulement les colonnes num√©riques
num_cols = ["prix", "quantit√©_command√©e", "stock_minimum", "niveau_stock"]
cat_cols = [col for col in X_encoded.columns if col not in num_cols]

# Normalisation
scaler = StandardScaler()
X_encoded_scaled = X_encoded.copy()
X_encoded_scaled[num_cols] = scaler.fit_transform(X_encoded_scaled[num_cols])
```

```python
# Split identique (m√™me random_state)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

### üìà √âtape 2 : Entra√Ænement Logistic Regression
On utilise `max_iter=1000` pour √©viter les avertissements de convergence.


```python
from sklearn.linear_model import LogisticRegression

# Cr√©ation et entra√Ænement
logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(X_train, y_train)
```

```python
# Pr√©diction
y_pred_logreg = logreg.predict(X_test)

# R√©sultats sous forme de DataFrame
report_logreg = classification_report(y_test, y_pred_logreg, output_dict=True)
df_logreg = pd.DataFrame(report_logreg).T
display(df_logreg.round(2))

# Matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred_logreg)
plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Purples", xticklabels=["Pas Retour", "Retour"], yticklabels=["Pas Retour", "Retour"])
plt.title("Matrice de Confusion (Logistic Regression)")
plt.xlabel("Pr√©dit")
plt.ylabel("R√©el")
plt.show()
```

### üìâ √âtape : Courbe ROC & AUC
- La **courbe ROC** montre la performance du mod√®le √† tous les seuils de classification possibles.
- L‚Äô**AUC (Area Under Curve)** mesure la capacit√© globale du mod√®le √† s√©parer les classes.
  - **AUC = 1.00** : parfait
  - **AUC = 0.5** : nul (comme du hasard)


```python
from sklearn.metrics import roc_curve, roc_auc_score

# üî¢ Probabilit√© pr√©dite pour la classe "1" (retour)
y_proba_logreg = logreg.predict_proba(X_test)[:, 1]

# üìà ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba_logreg)
auc_score = roc_auc_score(y_test, y_proba_logreg)

# üé® Trac√©
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"ROC curve (AUC = {auc_score:.2f})", color="darkorange")
plt.plot([0, 1], [0, 1], "k--", label="Al√©atoire")
plt.xlabel("Taux de faux positifs (FPR)")
plt.ylabel("Taux de vrais positifs (TPR)")
plt.title("Courbe ROC ‚Äì Logistic Regression")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

## Algorithme de ml XGBoost

### üì¶ √âtape 1 : Entra√Ænement du mod√®le XGBoost
On commence par cr√©er un mod√®le **XGBoost** avec la classe `XGBClassifier`. Ensuite, on l‚Äôentra√Æne sur les donn√©es d‚Äôentra√Ænement (`X_train`, `y_train`).


```python
import xgboost as xgb

# üåü Cr√©ation et entra√Ænement du mod√®le XGBoost
xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)
```

### üî¢ √âtape 2 : Pr√©dictions
Apr√®s l‚Äôentra√Ænement, le mod√®le pr√©dit les **classes** (`y_pred_xgb`) et les **probabilit√©s** (`y_proba_xgb`) pour chaque √©chantillon dans le jeu de test.
Les probabilit√©s de la classe "1" (retour) sont extraites pour tracer la courbe ROC.


```python
# üî¢ Pr√©dictions (classes et probabilit√©s)
y_pred_xgb = xgb_model.predict(X_test)
y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]
```

### üßæ √âtape 3 : Rapport de classification
On g√©n√®re un **classification report** qui r√©sume la performance du mod√®le : pr√©cision, rappel, F1-score et autres m√©triques.
Il est transform√© en DataFrame pour un affichage plus lisible.


```python
from sklearn.metrics import classification_report

# üßæ Classification report
report_xgb = classification_report(y_test, y_pred_xgb, output_dict=True)
df_xgb = pd.DataFrame(report_xgb).T
display(df_xgb.round(2))
```

### üìä √âtape 4 : Matrice de confusion
La **matrice de confusion** montre combien de commandes ont √©t√© correctement ou incorrectement pr√©dites comme retourn√©es ou non retourn√©es.
Cela permet de visualiser **les faux positifs et les faux n√©gatifs**.


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# üìä Matrice de confusion
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix_xgb, annot=True, fmt="d", cmap="Purples", xticklabels=["Pas Retour", "Retour"], yticklabels=["Pas Retour", "Retour"])
plt.title("Matrice de Confusion (XGBoost)")
plt.xlabel("Pr√©dit")
plt.ylabel("R√©el")
plt.show()
```

### üìà √âtape 5 : Courbe ROC + AUC
La **courbe ROC** et l‚Äô**AUC** permettent d‚Äô√©valuer la qualit√© du mod√®le pour s√©parer les classes (retour vs pas retour).
Plus l'AUC est √©lev√©, meilleure est la capacit√© de pr√©diction du mod√®le.


```python
from sklearn.metrics import roc_curve, roc_auc_score

# üìà Courbe ROC + AUC
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_proba_xgb)
auc_xgb = roc_auc_score(y_test, y_proba_xgb)

# üé® Trac√© de la courbe ROC
plt.figure(figsize=(6, 5))
plt.plot(fpr_xgb, tpr_xgb, label=f"XGBoost (AUC = {auc_xgb:.2f})", color="blue")
plt.plot([0, 1], [0, 1], "k--", label="Al√©atoire")
plt.xlabel("Taux de faux positifs (FPR)")
plt.ylabel("Taux de vrais positifs (TPR)")
plt.title("Courbe ROC ‚Äì XGBoost")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

Ce graphique compare les **courbes ROC** des trois mod√®les :

- L‚Äô**AUC** (area under curve) de chaque mod√®le est affich√©e dans la l√©gende.
