# Indicateurs clés de performances
- merge produits-commandes-expéditions-retours-stocks

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import plotly.express as px
import plotly.graph_objects as go 
```

```python
# Chargement des données
produits = pd.read_csv('../base/produitsV2.csv',sep=";")
commandes = pd.read_csv('../base/commandesV2.csv',sep=";")
expeditions = pd.read_csv('../base/expeditionsV2.csv',sep=";")
retours = pd.read_csv('../base/retoursV2.csv',sep=";") 
stocks = pd.read_csv('../base/stocksV2.csv',sep=";")
```

```python

# Fusion progressive
df_scenario1 = produits.merge(
    commandes, 
    left_on='id_produit', 
    right_on='id_produit', 
    how='left'
).merge(
    expeditions,
    left_on='id_commande',
    right_on='id_commande',
    how='left',
    suffixes=('_prod', '_cmd')
).merge(
    retours,
    left_on='id_commande',
    right_on='id_commande',
    how='left'
).merge(
    stocks,
    left_on='id_produit',
    right_on='id_produit',
    how='left'
)
```

```python
# Nettoyage des colonnes
cols_to_drop = ['entrepôt_cmd', 'id_produit_y']
df_scenario1 = df_scenario1.drop(columns=[c for c in cols_to_drop if c in df_scenario1.columns])

# Renommage explicite
df_scenario1 = df_scenario1.rename(columns={
    'entrepôt_x': 'entrepôt_stock',
    'quantité_x': 'quantité_commandée'
})

# Export
df_scenario1.to_csv('consolidation_cycle_vie.csv', index=False)
```
