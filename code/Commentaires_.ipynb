## Méthodologie et étapes du projet

### Phase 1 : Préparation et compréhension des données

La première étape consiste à établir une vision claire des relations entre les différentes tables de données qui constituent notre écosystème d'informations.

- **Documentation détaillée des variables** avec un dictionnaire de données précisant le type, la signification et les relations
- **Identification des clés primaires et étrangères** reliant les tables entre elles
- **Exploration initiale de la qualité des données** pour identifier les valeurs manquantes, aberrantes ou incohérentes

L'analyse des fichiers disponibles révèle déjà une structure de données riche comprenant des informations sur les produits (1000 références), les commandes, les expéditions, les retours, les stocks et les fournisseurs[1][2][3][4][6].

### Phase 2 : Nettoyage et préprocessing des données

Cette phase est critique pour garantir la fiabilité des analyses ultérieures :

## Indicateurs clés de performance (KPI) pour la chaîne d'approvisionnement

Plusieurs indicateurs spécifiques seront calculés et analysés :

### KPI de gestion des stocks

- **Taux de rotation des stocks** = Consommation annuelle / Stock moyen
- **Jours de couverture de stock** = (Niveau de stock / Consommation moyenne journalière)
- **Taux de rupture de stock** = Nombre de jours en rupture / Nombre total de jours
- **Écart entre stock réel et stock minimum** par référence[2]

### KPI d'expédition et livraison

- **OTIF (On-Time In-Full)** = % des livraisons complètes et à l'heure
- **Délai moyen d'expédition** par transporteur et région
- **Coût logistique unitaire** = Coût total de transport / Nombre d'unités livrées
- **Fill Rate** = Quantité livrée à la commande / Quantité commandée × 100[9]

### KPI de qualité et satisfaction

- **Taux de retour** par catégorie de produit et motif
- **Distribution des motifs de retour** (produit abîmé, mauvaise référence, etc.)
- **Indice de concentration HHI** pour mesurer la dépendance aux fournisseurs ou transporteurs[9]

L'analyse initiale des retours montre que les principaux motifs sont "Produit abîmé", "Livraison incomplète", "Mauvaise référence" et "Client insatisfait"[4].

## Optimisation Conjointe QCD par Programmation Linéaire

### Formalisation Mathématique
Objectif : 
$$ \min \sum_{t,w} (C_{stock} \cdot I_{t,w} + C_{penurie} \cdot S_{t,w}) $$
Contraintes :
$$ I_{t,w} = I_{t-1,w} + P_{t,w} - D_{t,w} $$
$$ D_{t,w} \geq 0.95 \cdot \hat{D}_{t,w} $$
$$ P_{t,w} \leq Capa_{w} $$

Où :
- $$ C_{stock} $$ = 15€/unité/semaine (coût de possession)
- $$ C_{penurie} $$ = 50€/unité (coût de rupture)
