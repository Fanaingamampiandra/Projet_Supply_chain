```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Chargement du fichier CSV
df = pd.read_csv("../base/achatsV2.csv", sep=';')
df
```

```python
# Conversion de la colonne date
df['date_achat'] = pd.to_datetime(df['date_achat'], dayfirst=True)
df
```

```python
# Création de nouvelles variables
df['montant_total'] = df['quantité'] * df['prix_unitaire']
df
```

```python
# Extraction de caractéristiques temporelles
df['mois'] = df['date_achat'].dt.month
df['année'] = df['date_achat'].dt.year
df['jour_semaine'] = df['date_achat'].dt.dayofweek  # 0=lundi
df
```

```python
# Vérification des valeurs manquantes
missing = df.isnull().sum()
df
```

```python
# Encodage des variables catégorielles (id_produit, id_fournisseur)
df_encoded = pd.get_dummies(df, columns=['id_produit', 'id_fournisseur'], drop_first=True)
df
```

```python
# Normalisation des variables numériques
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols_to_scale = ['quantité', 'prix_unitaire', 'délai_livraison_jours', 'montant_total']
df_encoded[cols_to_scale] = scaler.fit_transform(df_encoded[cols_to_scale])
df
```

```python
# Aperçu du résultat
print(df_encoded.head())
```

```python
import matplotlib.pyplot as plt

# Recréer la colonne montant_total
df['montant_total'] = df['quantité'] * df['prix_unitaire']

# 1. Répartition des montants totaux
df['montant_total'].plot.hist(bins=50, edgecolor='black', title='Répartition des montants totaux')
plt.xlabel("Montant total (€)")
plt.grid(True)
plt.show()

```

Une large concentration des achats entre 2 000 € et 12 000 €.

Une queue étendue vers la droite, indiquant quelques achats avec des montants très élevés (jusqu’à près de 30 000 €).

Le budget est majoritairement alloué à des achats de taille moyenne.

Une minorité de transactions très coûteuses peut représenter un risque ou une opportunité à analyser plus en détai

```python
# 2. Top 10 produits les plus achetés (par fréquence)
top_produits = df['id_produit'].value_counts().head(10)
top_produits.plot(kind='bar', title='Top 10 produits les plus achetés')
plt.ylabel("Nombre d'achats")
plt.xlabel("ID Produit")
plt.grid(axis='y')
plt.show()
```

Ces produits apparaissent le plus souvent dans les achats (pas nécessairement les plus coûteux).

Cela peut être dû à une forte demande, une consommation régulière ou leur rôle essentiel dans l’activité.

Ces produits peuvent être stratégiques : surveiller leur stock, leur disponibilité et les relations avec leurs fournisseurs.

```python
# 3. Top 10 fournisseurs les plus fréquents
top_fournisseurs = df['id_fournisseur'].value_counts().head(10)
top_fournisseurs.plot(kind='bar', color='orange', title='Top 10 fournisseurs les plus utilisés')
plt.ylabel("Nombre d'achats")
plt.xlabel("ID Fournisseur")
plt.grid(axis='y')
plt.show()
```

Certains fournisseurs comme S024, S012, etc., dominent largement les achats.

Peut-être liés à la diversité de leur catalogue, leur fiabilité ou leur capacité de livraison.

Cela peut indiquer une dépendance à certains fournisseurs : il est utile d’évaluer le risque ou négocier des conditions plus avantageuses.

```python
# 4. Évolution du montant total des achats dans le temps (mensuel)
df_monthly = df.set_index('date_achat').resample('M')['montant_total'].sum()
df_monthly.plot(title="Évolution mensuelle des montants totaux")
plt.ylabel("Montant total (€)")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

Des pics ou creux dans certains mois.

La tendance permet d’identifier des saisonnalités, des périodes de forte activité ou de ralentissement.

Peut aider à planifier les budgets ou optimiser les ressources logistiques selon les périodes de l’année.

```python
# Calculer le coût total par produit
produits_couteux = df.groupby('id_produit')['montant_total'].sum().sort_values(ascending=False).head(10)

# Afficher en graphique
produits_couteux.plot(kind='bar', title='Top 10 des produits les plus coûteux (montant total des achats)')
plt.ylabel("Montant total (€)")
plt.xlabel("ID Produit")
plt.grid(axis='y')
plt.show()
```

Ces produits ont généré les plus grandes dépenses cumulées.

Cela peut résulter d’un prix unitaire élevé ou de grandes quantités achetées.

Ce sont des produits à fort enjeu financier.

Il est pertinent d’analyser leur rentabilité, leur impact stratégique, ou de chercher à optimiser leurs coûts.
